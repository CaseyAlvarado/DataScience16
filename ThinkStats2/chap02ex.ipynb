```python
print "mean for parities:", np.mean(parities)

##from thinkstats2 code:
def Var(xs, mu=None, ddof=0):
    """Computes variance.

    xs: sequence of values
    mu: option known mean
    ddof: delta degrees of freedom

    returns: float
    """
    xs = np.asarray(xs)

    if mu is None:
        mu = xs.mean()

    ds = xs - mu
    return np.dot(ds, ds) / (len(xs) - ddof)

def Std(xs, mu=None, ddof=0):
    """Computes standard deviation.

    xs: sequence of values
    mu: option known mean
    ddof: delta degrees of freedom

    returns: float
    """
    var = Var(xs, mu, ddof)
    return math.sqrt(var)

print "Variance for parities:", Var(parities)
print "Standard deviation for parities:", Std(parities)
```

### Exercise 4

Using the variable totalwgt_lb, investigate whether first babies are lighter or heavier than others. Compute Cohen’s d to quantify the difference between the groups. How does it compare to the difference in pregnancy length? 

```python
import nsfg
pregs = nsfg.ReadFemPreg()
```

```python
firsts = pregs[pregs.birthord == 1]
others = pregs[pregs.birthord != 1]
```

```python
first_weights = firsts.totalwgt_lb.value_counts()
other_weights = others.totalwgt_lb.value_counts()
print "mean total weight of first born babies:", first_weights.mean()
print "mean total weight of other born babies:", other_weights.mean()
```

```python
def Cohen(one, two): 
    diff = one.mean() - two.mean()
    
    var1 = one.var()
    var2 = two.var()
    n1, n2 = len(one), len(two)
    
    pooled_var = (n1 *var1 + n2*var2)/(n1 + n2)
    d = diff / math.sqrt(pooled_var)
    return math.fabs(d)

print Cohen( first_weights, other_weights)
```

The difference between the two total weights is 0.08667245291558709 standard deviations away from each other. This number is incredibly small. This means that the total weight of the first born child is not too different from the weight of the other children, their weights are relatively close together when born.

```python
firsts_length = firsts.prglngth
others_length = others.prglngth
print "mean pregnancy length of first born babies:", firsts_length.mean()
print "mean pregnancy length weight of other born babies:", others_length.mean()
```

```python
Cohen(firsts_length, others_length)
```

According to these calculations, women are pregnant with their first born children longer than they are pregnant with other children by a significant amount. Oddly enough this was not the conclusion Allen came to in the book. With this data, women are pregnant around 13.4297099034 weeks longer with their first born than with any of the later children. Compared to total weights, there is a significant difference in pregnancy length between firsts and others than there was when comparing total weights. Having a medium Cohen value means that the strength of this phenomenon is medium, meaning that it has a medium effect or medium correlation betweeen the two pregnancy lengths
